```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from xgboost import XGBRegressor
```

```python
train_o = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv', index_col="Id")
test_o = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv', index_col="Id")
trainx = train_o.iloc[:,:-1]
trainy = train_o.iloc[:,-1]
```

```python
cat = [i for i in trainx if not (trainx[i].dtype == "int64" or trainx[i].dtype == "float64")]
```

```python
col_trans = ColumnTransformer([('cat', OneHotEncoder(handle_unknown='ignore'), cat)], remainder='passthrough')

xgbreg = XGBRegressor()

pipeline = make_pipeline(col_trans, xgbreg)
```

```python
cross_val_score(pipeline, train_o.iloc[:,:-1], train_o.iloc[:,-1], cv=5)
```

```python
params = {
    "xgbregressor__gamma": np.arange(0,150,5),
    "xgbregressor__learning_rate": np.arange(0.1,0.6, 0.1),
    "xgbregressor__n_estimators": np.arange(1,601,20),
    "xgbregressor__reg_alpha": np.arange(0,0.4,0.05),
    "xgbregressor__reg_lambda": np.arange(1,5,2)
}

gridsearch = GridSearchCV(pipeline, params)
gridsearch.fit(trainx, trainy)
```
